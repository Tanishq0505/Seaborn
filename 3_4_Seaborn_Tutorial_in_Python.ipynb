Seaborn (sns): Seaborn is a popular data visualization library in Python that is based on Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.



Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

load_dataset() function: Seaborn comes with several built-in datasets that can be used for practice and examples. The load_dataset() function is used to load these datasets.

'tips' dataset: The 'tips' dataset is one of the built-in datasets in Seaborn. It contains information about tips given in a restaurant, including variables like the total bill, tip amount, sex of the bill payer, whether they are a smoker, the day of the week, the time of day, and the size of the party.



```python
# total bill vs tip dataset
tips = sns.load_dataset('tips') #The code tips = sns.load_dataset('tips') is used in Python to load a sample dataset 
# called "tips" using the Seaborn library. 
# tips: This variable is assigned the DataFrame that contains the 'tips' dataset.
```

```python
tips.head()
```

```python
# setting a theme for the plots

sns.set_theme() #The code sns.set_theme() is used to set the visual theme or style for Seaborn plots. Seaborn provides 
# several themes that change the appearance of the plots, making it easy to customize the look and feel of the visualizations.
```

Explaination below code -> sns.relplot(data=tips, ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')

The code sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size') is used to create a relational plot using Seaborn's relplot function. This function is very flexible and allows you to visualize the relationships between multiple variables in a dataset. Let's break down what each part of this code does:

sns.relplot(): This is the function from the Seaborn library that creates a relational plot. It can be used to generate scatter plots or line plots that show the relationship between two variables.

data=tips: This parameter specifies the dataset to be used for the plot. Here, the dataset is tips, which contains information about restaurant bills, tips, and other related variables.

x='total_bill': This parameter sets the variable for the x-axis. In this case, it is the 'total_bill' column from the tips dataset, representing the total bill amount.

y='tip': This parameter sets the variable for the y-axis. Here, it is the 'tip' column from the tips dataset, representing the tip amount.

col='time': This parameter creates separate 2 subplots (facets) for each unique value of the 'time' variable. In the tips dataset, 'time' typically represents whether the meal was lunch or dinner. This means the data will be split into two plots, one for time=lunch and one for time=dinner.

hue='smoker': This parameter colors the points based on the 'smoker' variable, which indicates whether the bill payer is a smoker or not. Different colors will be used for smokers and non-smokers.

style='smoker': This parameter changes the style (marker shape) of the points based on the 'smoker' variable. This means smokers and non-smokers will have different marker shapes.

size='size': This parameter scales the size of the points based on the 'size' variable, which represents the size of the dining party. Larger parties will have larger points in the plot.



```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# importing the california house price data
from sklearn.datasets import fetch_california_housing

#Now  Load the California housing dataset to the variable california_data

# Below line calls the fetch_california_housing function and assigns the returned value to the variable california_data.
california_data = fetch_california_housing()
```

Distribution Plot

```python
# Create a histogram using seaborn's histplot

# Creates a histogram plot of the 'Population' column from the california_df DataFrame using Seaborn's histplot function.
sns.histplot(california_df['Population'], kde=True)  
#kde=True: Specifies to add a kernel density estimate (KDE) to the histogram plot, which provides a smoothed curve representing the distribution of the data.


plt.title('Distribution of Population in California')  # Set the plot title
plt.xlabel('Population')  # Set the x-axis label
plt.ylabel('count')  # Set the y-axis label
plt.show()  # Display the plot
```

Correlation:
- Positive Correlation
- Negative Correlation

.corr() is a method provided by pandas DataFrames.

It computes pairwise correlation of columns, excluding NA/null values.

By default, it computes the Pearson correlation coefficient between all pairs of numerical columns in the DataFrame.

The Pearson correlation coefficient measures the linear relationship between two variables. It ranges from -1 to +1:

+1: Total positive linear correlation.

0: No linear correlation.

-1: Total negative linear correlation.


```python
correlation = california_df.corr()
california_df.corr()
```

Heat Map 

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')

# cbar=True: This adds a color bar on the right side of the heatmap. The color bar is a visual guide that shows the mapping of colors to correlation values.
# square=True: This ensures that each cell in the heatmap is square in shape. Without this, the cells could be rectangular.
# fmt='.1f': This formats the annotation text inside the cells. '.1f' means that the values inside the cells will be displayed as floating-point numbers with one decimal place.
# annot=True: This adds the actual correlation values as annotations in each cell. It ensures that the correlation numbers are displayed within the 
# heatmap squares.
# annot_kws={'size':8}: This controls the size of the annotation text inside the cells. In this case, the font size is set to 8.
# cmap='Blues': This sets the color map for the heatmap. 'Blues' is a gradient from light blue (for low correlation values) to dark blue 
# (for high correlation values). You can change the color map to other options like 'coolwarm', 'viridis', etc., depending on your preference.




```
